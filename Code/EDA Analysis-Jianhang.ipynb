```python
# Convert 'salary_pay' column to numeric values
df['salary_pay'] = pd.to_numeric(df['salary_pay'], errors='coerce')
```

```python
df['salary_pay'].mean()
```

```python
# Check the sample size. since there are a lot of missing values, find the percent of jobs with salaries Pay
total_jobs = len(df)
jobs_with_salaries = df['salary_pay'].count()

percent_jobs_with_salaries = (jobs_with_salaries / total_jobs) * 100
percent_jobs_with_salaries
```

```python
# Check for outliers in the salary by creating a box plot
sns.set(style="whitegrid")  

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['salary_pay'])

plt.xlabel('Salary Pay')
plt.title('Box Plot of Salary Pay with Outliers')

plt.show()
```

```python
# Check the sample size. since there are a lot of missing values, find the percent of jobs with salary standarized
total_jobs = len(df)
jobs_with_salaries = df['salary_standardized'].count()

percent_jobs_with_salaries = (jobs_with_salaries / total_jobs) * 100
percent_jobs_with_salaries
```

```python
# Check for outliers in the salary by creating a box plot
sns.set(style="whitegrid")  

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['salary_standardized'])

plt.xlabel('Salary Standarized')
plt.title('Box Plot of Salary Standarized with Outliers')

plt.show()
```

```python
# Group the data by job title and calculate the average salary for each title
avg_salary_by_title = df.groupby('title')['salary_avg'].mean()

# Sort the data in descending order and select the top 10
top_10 = avg_salary_by_title.sort_values(ascending=False).head(10)


plt.figure(figsize=(12,8))
plt.bar(top_10.index, top_10.values)
plt.xticks(rotation=90)
plt.xlabel('Job Title')
plt.ylabel('Average Salary ($)')
plt.title('Top 10 Job Titles with the Highest Average Salary')
plt.show()
```

```python
jobs_by_location = df.groupby('location')['title'].count()
top_10 = jobs_by_location.sort_values(ascending=False).head(10)
plt.figure(figsize=(12,8))
plt.bar(top_10.index, top_10.values)
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Number of Job Postings')
plt.title('Distribution of Data Analyst Job Postings by Location')
plt.show()
```

```python
unique_states=set(item for item in df["location"]) #look for unique states
print(unique_states)
```

```python
state_code=[]
for place in unique_states :
    if isinstance(place, str):
        parts=place.split(",")
        if len(parts)>1:
            state_codes=parts[1].strip()
            state_code.append(state_codes)

print(state_code)
```

```python
#check what skills are the most demanding
skills_python=df[df['description_tokens'].str.contains('python')]
skills_sql=df[df['description_tokens'].str.contains('sql')]
skills_tableau=df[df['description_tokens'].str.contains('tableau')]
skills_power_bi=df[df['description_tokens'].str.contains('power_bi')]
python_frequency=skills_python.shape[0]
sql_frequency=skills_sql.shape[0]
tableau_frequency=skills_tableau.shape[0]
power_bi_frequency=skills_power_bi.shape[0]
plt.bar(['Python','SQL','Tableau','Power_BI'], [python_frequency, sql_frequency,tableau_frequency,power_bi_frequency])
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.title('Frequency of Skills in Job Descriptions')
plt.show()
```
