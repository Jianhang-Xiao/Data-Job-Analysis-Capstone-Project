```python
# Group the data by job title and calculate the average salary for each title
avg_salary_by_title = df.groupby('title')['salary_avg'].mean()

# Sort the data in descending order and select the top 10
top_10 = avg_salary_by_title.sort_values(ascending=False).head(10)


plt.figure(figsize=(12,8))
plt.bar(top_10.index, top_10.values)
plt.xticks(rotation=90)
plt.xlabel('Job Title')
plt.ylabel('Average Salary ($)')
plt.title('Top 10 Job Titles with the Highest Average Salary')
plt.show()
```

```python
unique_states=set(item for item in df["location"]) #look for unique states
print(unique_states)
```

```python
state_code=[]
for place in unique_states :
    if isinstance(place, str):
        parts=place.split(",")
        if len(parts)>1:
            state_codes=parts[1].strip()
            state_code.append(state_codes)

print(state_code)
```

```python
#check what skills are the most demanding
skills_python=df[df['description_tokens'].str.contains('python')]
skills_sql=df[df['description_tokens'].str.contains('sql')]
skills_tableau=df[df['description_tokens'].str.contains('tableau')]
skills_power_bi=df[df['description_tokens'].str.contains('power_bi')]
python_frequency=skills_python.shape[0]
sql_frequency=skills_sql.shape[0]
tableau_frequency=skills_tableau.shape[0]
power_bi_frequency=skills_power_bi.shape[0]
plt.bar(['Python','SQL','Tableau','Power_BI'], [python_frequency, sql_frequency,tableau_frequency,power_bi_frequency])
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.title('Frequency of Skills in Job Descriptions')
plt.show()
```
